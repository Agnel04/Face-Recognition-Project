# Main part

```python
def face_encoding(img) :

    encoded_list = []
    
    for i in imgs :

        img = cv2.cvtColor(i,cv2.COLOR_BGR2RGB)

        face_loc = fr.face_locations(img)
        # print(face_loc)
        face_enc = fr.face_encodings(img, face_loc)[0]
        # print(face_enc)

        encoded_list.append(face_enc)
    
    return encoded_list
```

```python
trained_images = face_encoding(imgs)
trained_images
```

```python
#  opening camera

video = cv2.VideoCapture(0)

while True :

    success, img = video.read()
    img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    face_in_frame = fr.face_locations(img1)
    # print(face_in_frame)
    face_encoder = fr.face_encodings(img1, face_in_frame)   #not taking from 0th index for allowing it to detect multiple faces
    # print(face_encoder)

    for encoder, faceloc in zip(face_encoder, face_in_frame) :

        match = fr.compare_faces(trained_images, encoder)       # -> this method may detect 1 person as 2 or more people..to overcome this we go for next method 
        # print(match)      

        # face distance
        distance = fr.face_distance(trained_images, encoder)    # -> this method gives the distance b/w the faces.. the least distance will be the correct match
        index = np.argmin(distance)
        # print(index)

        # printing name of the person
        if match[index] :
            name = clsnames[index]
            print(name)
        else :
            name = 'Unknown Person'
            print(name)

        y1,x2,y2,x1 = faceloc
        cv2.rectangle(img, (x1,y1), (x2,y2), (0,0,255),2)
        cv2.putText(img, name, (x1,y1), cv2.FONT_HERSHEY_COMPLEX, 2, (0,255,0),2)

    cv2.imshow('video', img)
    if cv2.waitKey(1) & 0XFF == ord('q') :  # video stops when the key 'q' is pressed
        break

video.release()
cv2.destroyAllWindows()
```
